```python
# Question 1 FDA Assignment:
import pandas
import numpy

def Interevent_Calculation():

  # enter the array so that we can split the values by comma
  a = input('Enter the 1 dimension list or array with 1 space in between: ')
  n_list = list(map(int, a.split(' ')))
  ip_array = numpy.array(n_list)

  #Enter Threshold:
  Thresh = int(input('Please set the threshold value: '))

  # Converting Array to DF
  df = pandas.DataFrame(ip_array)
  df = df.reset_index()

  # drop values where i > threshold
  df_final = df[df[0] > Thresh]

  Interevent_Times = df_final['index'] - df_final['index'].shift(+1)
  Interevent_Times = Interevent_Times.dropna()

  # Converting data_frame into  array again
  Interevent_Times = numpy.array(Interevent_Times)

  return Interevent_Times

Interevent_Calculation()
```

```python
#  Question 2 FDA Assignment:
import numpy 

a = input('Enter the 1 dimension list or array with a space in between: ')
a_list = list(map(int, a.split(' ')))
ip_array = numpy.array(a_list)

def adjacency_matrix(data):

    dim = len(data)

    A = numpy.zeros((dim,dim))

    for c in range(dim):
        if c<dim-1:
            A[c,c+1] = A[c+1,c] = 1

        for j in range(c + 2, dim):
            if numpy.max(data[c+1:j])<=min(data[c], data[c]):
                A[c,j] = A[j,c] = 1

    return A

adjacency_matrix(ip_array)
```
